# Tipi di dato

In Python tutto è un oggetto e ogni oggetto ha un tipo. Esistono diversi modi per classificare i tipi di dato. Ad esempio, puoi dividerli in scalari, che non possono essere suddivisi ulteriormente (come un numero), e non scalari, che possono essere suddivisi (come una lista di numeri).

Un altro modo per classificarli è in base alla loro origine:
- Tipi di dato integrati (Built-in Data Types)
- Tipi di dato non integrati nell’interprete


## Booleani

I booleani sono un altro tipo di dato fondamentale in Python e consistono solo in due valori: **True** e **False**. I valori booleani vengono usati principalmente quando confrontiamo diversi valori tra loro (ne parleremo meglio più avanti).
In Python, True ha valore 1 mentre False ha valore 0. Possiamo verificare questo fatto convertendoli in interi:

```python
print(int(True))
print(int(False))
```

### Conversione (cast) tra interi e float

Usando **float( )** e **int( )** possiamo convertire gli interi in float e viceversa:

```python
val = 34.67
type(val)
int(val)
print(val)
```

```python
print(int(45.89)) # convertire un float in int
print(int('347')) # convertire una stringa in int
```

```python
print(int('56.76')) # questo non funziona
```

```python
print(int(float('56.76'))) # questo sì...perché?
```

```python
#Lo stesso non vale per la conversione da stringa a float.
#Significa che la conversione funziona solo se la stringa ha la forma di un float.

float('4')
```

## Controllare i tipi di dato

Ci sono due modi principali per verificare il tipo di dato di un oggetto. Il più semplice è usare **type**, come abbiamo visto all’inizio di questo notebook:

```python
print(type(4524))
print(type('a very good dog'))
```

Usando questo metodo, possiamo verificare se una variabile ha un certo tipo:

```python
type([3, [], {'foo': 'bar'}]) == dict
```

Un modo più “Pythonic” per farlo è utilizzare la funzione incorporata **.isinstance:**

```python
isinstance('49.890', str) # Verificare se ‘49.890’ è una stringa
```

Il principale vantaggio di questa funzione rispetto a **type** è che possiamo usarla per effettuare verifiche multiple:

```python
age_inputs = [11, 14, 13, '12', 12, [16]]

for age in age_inputs:
    if not isinstance(age, (int, str) ):
        print(f'{age} is {type(age)}. It should be either integer or string')
```
