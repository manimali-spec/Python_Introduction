# Functions

Le funzioni in Python sono un insieme di codice che eseguono una funzionalità specifica se questa funzionalità deve essere riutilizzata in seguito in un’altra parte del codice.

Supponiamo di voler eseguire alcune statistiche semplici su una lista di numeri:

```python
import numpy as np
np.random.seed(42)

data = np.random.rand(1,100_000).tolist()[0]
std = np.std(data)
mean = np.mean(data)
median = np.median(data)
min_val = min(data)
max_val = max(data)
for val in [std, mean, median, min_val, max_val]:
    print(val)
```

Ok, l’abbiamo fatto! Ma cosa succede se ora vogliamo ripetere questi processi per un’altra lista di numeri? Abbiamo due opzioni:
- Aggiornare la variabile dei dati con una nuova lista
- Riscrivere tutti i passaggi per un’altra lista

Entrambe queste soluzioni sono sbagliate! O almeno non sono Pythoniche. Il modo corretto di affrontare questo compito è racchiudere i passaggi in una funzione.

## Built-in functions

Fino a questo punto, abbiamo utilizzato molte funzioni senza conoscere la definizione esatta di una funzione! Abbiamo visto:
- print
- len
- max
- min

Questi sono solo alcuni esempi tra le centinaia di funzioni che fanno parte del linguaggio Python di base (significa che per usarle non è necessario importare alcuna libreria o pacchetto). Una funzione ha sempre la seguente struttura:

I parametri sono input che le funzioni possono richiedere o meno per svolgere il compito desiderato. Ad esempio, diamo un’occhiata alla funzione print che prende una stringa come input e la stampa.

```python
print('My shiny text!')
print() # prints an empty line
print('X', 'Y', 'Z', sep='+++', end='\n')
print('X', 'Y', 'Z', '+++', '°°°')
```

## Funzioni personalizzate

Anche se esistono numerosi moduli e pacchetti incorporati o di terze parti pronti all’uso, ci sono alcuni compiti troppo specifici per i quali non esiste alcun modulo o funzione già pronta che possa svolgerli. In questi casi (che accadono spesso) dobbiamo scrivere le nostre funzioni personalizzate.
Considera l’esempio iniziale che abbiamo visto all’inizio della sezione “funzioni”, cioè il calcolo di alcune semplici statistiche.

```python
def simple_stat(data, log=True):
    """Calculating simple statistics of a given list.
    
    For the the given list of numbers, calculates standard deviation, 
    Mean, Median, Minimum and Maximum statistics.

    Args:
        data : A list of numerical data.
        log: If True, prints a simple report.
        
    Returns:
        Return a list of the given numbers
    """
    std = np.std(data)
    mean = np.mean(data)
    median = np.median(data)
    min_val = min(data)
    max_val = max(data)
    
    vals = [std, mean, median, min_val, max_val]
    names = ['Standard deviation', 'Mean', 'Median', 'Min', 'Max']
    
    if log:
        for name, val in zip(names, vals):
            print(f'{name} : {val}')
            #print(f'{name} : {round(val, 2)}')
    return vals
```

```python
def simple_stat(data):
    
    std = np.std(data)
    mean = np.mean(data)
    median = np.median(data)
    min_val = min(data)
    max_val = max(data)
    
    vals = [std, mean, median, min_val, max_val]
    names = ['Standard deviation', 'Mean', 'Median', 'Min', 'Max']
    
    for name, val in zip(names, vals):
        print(f'{name} : {val}')
    return vals
```
