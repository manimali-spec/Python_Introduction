Aspetta, cos’è invalid syntax ora! Otteniamo questo errore quando Python pensa che ciò che hai scritto non abbia senso! Per capire meglio cos’è un errore di sintassi, vediamo un esempio di linguaggio naturale:
- Questa è una frase corretta $\rightarrow$ Marco eats sandwich $\rightarrow$ (sostantivo) (verbo) (sostantivo)
- Questa è una frase sintatticamente sbagliata $\rightarrow$ Marco cat sandwich $\rightarrow$ (sostantivo) (sostantivo) (sostantivo)

Quindi, significa che qualsiasi combinazione (sostantivo) (verbo) (sostantivo) ci dà una combinazione corretta? Sì e no! Sì, perché (sostantivo) (verbo) (sostantivo) è una frase sintatticamente corretta, ma potrebbe comunque non avere senso $\rightarrow$ Marco ascolta il panino (ha un errore semantico statico).

La stessa cosa esiste in Python:
- ‘Seven’11 è sintatticamente sbagliato perché Python non capisce cosa intendi con (str)(int) $\rightarrow$ SyntaxError
- ‘seven’ * ‘two’ è anche sbagliato, ma non sintatticamente, perché anche se Python capisce quale sia il tuo obiettivo (moltiplicare due stringhe), non può farlo $\rightarrow$ TypeError

Tornando al nostro esempio nell’ultima cella, riesci a capire qual è il problema? (ovviamente puoi!)
Ci sono due soluzioni qui:
- Usare stringhe alternative come abbiamo fatto prima (’ ’ o “ “)
- Usare il carattere di escape: in Python (e in altri linguaggi di programmazione) alcuni caratteri hanno un ruolo riservato. Ciò significa che non tutti i caratteri in Python hanno la stessa caratteristica dietro le quinte. Per forzare Python a ignorare il ruolo speciale di alcuni caratteri come ' o ", possiamo usare una barra rovesciata:

```python
'I can\'t go to work today'
```

### Operazioni comuni sulle stringhe

Avendo una stringa, puoi applicare su di essa una vasta gamma di metodi per modificarla. Vedremo insieme le più comunemente usate:

#### Split

Restituisce una lista di pezzi della stringa originale tagliata sulla stringa indicata.

```python
text = 'This is a Python course'
print(text)
text = text.split(sep=' ')
print(text)
```

```python
text = 'I saw the guy, and suddenly he disappeared'
print(text)
text = text.split(',') # notice that I wrote ',' instead of sep=','
print(text)
```

#### Slicing

Tagliare la stringa in modo da ottenere una nuova stringa che sia un sottoinsieme della stringa originale.

In realtà, prima di spiegare lo slicing, dobbiamo imparare un concetto fondamentale in Python: l’indicizzazione (Indexing).
Supponiamo di avere una scatola con 11 penne dentro. Se vuoi prenderle una per una e contarle, probabilmente procederesti così:
- Uno
- Due
- …
- Undici

Questo significa che stai usando un indice basato su 1, indicando il primo oggetto con il numero 1. Python, invece, usa un indice basato su 0. Questo significa che si riferisce al primo oggetto come numero (indice) 0. Quindi, per contare le penne nel nostro esempio, Python va:
- Zero
- Uno
- …
- dieci

Ok, ora che sappiamo come funziona l’indicizzazione in Python, possiamo vedere l’indicizzazione delle stringhe.
In Python, l’indicizzazione/slicing non è limitata alle stringhe, ma possiamo fare slicing/indice su qualsiasi oggetto che possa essere considerato un gruppo di oggetti.
Detto ciò, dovrebbe essere chiaro ora che, ad esempio, la stringa: “cane”, pur essendo considerata un oggetto stringa, è anche vista come un contenitore (non come una definizione tecnica) di “c”, “a”, “n” e “e” con indici 0, 1, 2 e 3.
Questo significa che possiamo fare slicing su questa stringa utilizzando i loro indici. Vediamo un esempio:

```python
'New York'
```

```python
t = 'New York'
t[4:]
```

```python
text = 'NEW YORK'
print(text[1]) # E: second letter

print(text[0:3]) # NEW
print(text[40:80]) # YORK

print(text[:3]) # NEW
print(text[4:]) # YORK
```

Gli indici possono essere indirizzati non solo dall’inizio, ma anche dalla fine:

```python
text = 'NEW YORK'
print(text[-7]) # E: second letter

print(text[-8:-5]) # NEW
print(text[-4:]) # YORK

print(text[:-5]) # NEW
```

#### Replace

Restituisce una nuova versione della stringa originale in cui le istanze di str_to_be_replaced vengono sostituite con str_to_be_replaced_with:

```python
text = 'I love my dog and my little dog'
print(text.replace('dog', 'cat', 1))
```

```python
t = 'tavolo cavolo'
t.replace('cavolo', 'gatto')
```

#### Concatenation

La concatenazione ti consente di creare nuove stringhe concatenando insieme stringhe esistenti, utilizzando il simbolo “+”:

```python
text_1 = 'Hello'
text_2 = 'World!'
text_3 = text_1 + ' ' + text_2

print(text_3)
```

#### Strip

Rimuove la stringa data dall’inizio e dalla fine della stringa originale. Nota: Come la funzione split(), il valore predefinito della funzione strip() è lo spazio, quindi se esegui my_text.strip(), la funzione rimuove gli spazi iniziali e finali dalla stringa my_text.

```python
text_1 = ' Pirelli Spa '
print(text_1.strip())

text_2 = 'Continental -'
print(text_2.strip('-').strip())

text_3 = '*Python***'
print(text_3.lstrip('*')) # There is also "rsplit"
```
