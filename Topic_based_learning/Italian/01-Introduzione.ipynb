## Assegnazione dei nomi alle variabili

Ci sono 3 regole da tenere a mente quando si definisce una nuova variabile:
	•	Può contenere solo lettere, numeri e underscore (_).
	•	Deve iniziare con una lettera o un underscore (_). Non può iniziare con un numero.
	•	Non può / non dovrebbe essere un nome riservato.

Spieghiamo meglio l’ultimo punto. Ho usato **non può / non dovrebbe** e non solo **non può** perché, mentre è impossibile assegnare un valore ad alcune parole chiave riservate (come for), in realtà è possibile assegnare un valore ad alcune altre (come alla funzione print); il punto è che, anche se puoi farlo, è quasi sempre una pessima idea!

## Operatori di confronto

Gli operatori di confronto non sono tecnicamente considerati parte del “Control flow”, ma dato il loro utilizzo con le istruzioni menzionate, li includo qui. Come suggerisce il nome, questi operatori confrontano due oggetti e restituiscono **True** oppure **False**. Vediamo alcuni esempi:

```python
a = 15.6
b = 79
```

```python
print(a == b)
print(a != b)
print(a < b)
print(a > b)
print(a <= b)
print(a >= b)
```
