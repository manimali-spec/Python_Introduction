# Control Flow

Fino ad ora, abbiamo lavorato con elementi singoli. Per esempio, abbiamo fatto lo slicing di una stringa o scritto su un file. Ma cosa succede se vogliamo fare qualcosa (eseguire un’azione) su una sequenza di oggetti? Il primo tipo di ciclo che esploreremo è il ciclo for.

## For statement

Supponiamo di avere una lista con gli anni di nascita dei nostri colleghi e vogliamo calcolare le loro età, stamparle e inserirle in una nuova lista chiamata ages:

```python
years = [1987, 1990, 1980, 1973]
ages = [] # creating an empty list
```

Sappiamo come ottenere gli elementi di una lista usando gli indici. Quindi, cosa succede se prendiamo ciascun elemento e lo sottraiamo da 2025?

```python
age_1 = 2025 - years[0]
print(age_1)
ages.append(age_1)

age_2 = 2025 - years[1]
print(age_2)
ages.append(age_2)

age_3 = 2025 - years[2]
print(age_3)
ages.append(age_3)

age_4 = 2025 - years[3]
print(age_4)
ages.append(age_4)
```

```python
print(ages)
```

Ok, we did what we've we asked for! but I think you agree that it wasn't quite efficient as we've written more than 12 lines of code to perform three simple operations on 4 items! imagine you want to calculate the age of all Bicocca employees! Fortunately we for repetitive tasks like this( when we are dealing with sequence objects like lists, tuples, dictionaries or strings) we can use loops. In this case we're going to use a for loop:

```python
ages = []
for year in years:
    age = 2025 - year
    print(age)
    ages.append(age)
```

## If statement

Spesso, è necessario eseguire alcune istruzioni solo se una certa condizione è soddisfatta, oppure scegliere quali istruzioni eseguire tra diverse condizioni mutuamente esclusive.
L’istruzione composta di Python if, che utilizza le clausole if, elif ed else, permette di eseguire blocchi di codice in modo condizionato.
Ecco la sintassi dell’istruzione if:

È comune voler eseguire una parte di codice solo se una certa condizione è vera (True). Per farlo, in Python si usa l’istruzione if.
Torniamo al semplice esempio della stampa dei numeri che abbiamo visto con i cicli while e for, ma questa volta supponiamo di voler stampare il numero solo se è dispari:

```python
num = 0 # initial number is set to 0
while num <= 20: # while the number is less than or equal 20, continue
    if num%2 != 0: # if the reminder of them with 2
        print(num)
    num = num + 1 # adding one to the current value of the num = num+=1
```

Ok, now let's use also *eles* and *elif* clauses. Imagine that you have a set of ages and you want to print one if the following labels based on the age:
- kid
- Teen
- adult
- old

```python
ages = [56, 2, 14, 8, 33, 23, 19, 80, 6, 9, 27]

for age in ages:
    if age <= 12:
        print(age, 'Kid', sep=' - ')
    elif age <= 19 and age >= 13:
        print(age, 'Teen', sep=' - ')
    elif age <= 60 and age >= 20:
        print(f'{age} - Adult') # Did you noticed anything new here?
    else:
        print(age, 'Old', sep=' - ')
```
