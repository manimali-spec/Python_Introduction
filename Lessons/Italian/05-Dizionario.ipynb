## Dictionaries

I dizionari in Python sono strutture dati potenti utilizzate per memorizzare e gestire i dati in formato coppia chiave-valore. Ogni coppia chiave-valore mappa la chiave al suo valore associato. I dizionari sono mutabili, il che significa che possono essere modificati dopo la loro creazione. Sono indicizzati dalle chiavi, che devono essere tipi immutabili (ad esempio, stringhe, numeri o tuple con elementi immutabili) e unici all’interno di un dizionario.

Sintassi di base

Un dizionario in Python può essere creato utilizzando le parentesi graffe {} con coppie separate da virgole, e due punti : separano ogni chiave dal suo valore:

```python
my_dict = {'key1': 'value1', 'key2': 'value2'}
```

### Accesso agli Elementi

Puoi accedere a un valore facendo riferimento alla sua chiave tra parentesi quadre []:

```python
print(my_dict['key1'])  # Output: value1
```

Se provi ad accedere a una chiave che non esiste, verrà sollevato un KeyError. Per evitare questo, puoi usare il metodo **get()**, che restituisce None (o un valore predefinito specificato) se la chiave non viene trovata:

```python
print(my_dict.get('nonexistent_key'))
print(my_dict.get('nonexistent_key', 'default_value'))
```

### Aggiungere e Modificare Elementi

Puoi aggiungere una nuova coppia chiave-valore o modificare una esistente assegnando un valore a una chiave:

```python
my_dict['new_key'] = 'new_value'  # Adds a new key-value pair
my_dict['key1'] = 'updated_value'  # Updates the value of an existing key
```

### Rimuovere Elementi

Usa l’istruzione del per rimuovere una coppia chiave-valore specificando la sua chiave:

```python
del my_dict['key1']
```

Oppure, puoi usare il metodo **pop()**, che restituisce anche il valore associato alla chiave rimossa:

```python
removed_value = my_dict.pop('key2')
```

### Iterare su un Dizionario

Puoi iterare su un dizionario utilizzando un ciclo. Per impostazione predefinita, iterando su un dizionario si itera sulle sue chiavi:

```python
for key in my_dict:
    print(key, my_dict[key])
```

Puoi anche iterare su chiavi e valori insieme utilizzando il metodo **items()**:

```python
for key, value in my_dict.items():
    print(key, value)
```

### Comprensioni di Dizionari

Simile alle comprensioni delle liste, le comprensioni dei dizionari ti permettono di creare dizionari utilizzando una sintassi espressiva e concisa:

```python
squares = {x: x**2 for x in range(6)}
```

Questo crea un dizionario in cui ogni chiave è un numero da 0 a 5, e ogni valore è il quadrato della chiave.

### Metodi Utili per i Dizionari

- keys(): Restituisce un oggetto di vista contenente le chiavi del dizionario.
- values(): Restituisce un oggetto di vista contenente i valori del dizionario.
- items(): Restituisce un oggetto di vista contenente coppie (chiave, valore).
- clear(): Rimuove tutti gli elementi dal dizionario.
- copy(): Restituisce una copia superficiale del dizionario.
